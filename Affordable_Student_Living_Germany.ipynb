```python
import pandas as pd

df = pd.read_csv(r"C:\Users\abhi\Downloads\archive\cities_living_cost.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.columns
```

```python
german_cities = [
    'Berlin', 'Munich', 'Hamburg', 'Frankfurt', 'Cologne', 'Bonn',
    'Dusseldorf', 'Stuttgart', 'Leipzig', 'Dresden', 'Nuremberg',
    'Hannover', 'Bremen', 'Essen', 'Dortmund'
]
```

```python
df_germany = df[df['City'].isin(german_cities)].reset_index(drop=True)
df_germany
```

```python
df_germany.shape
```

```python
student_cols = [
    'City',
    'Cost_index',
    'Meal, Inexpensive Restaurant',
    'Monthly Pass (Regular Price)',
    'Apartment (1 bedroom) Outside of Centre',
    'Cinema, International Release, 1 Seat'
]

df_students = df_germany[student_cols]
df_students
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(df_students['City'], df_students['Cost_index'])
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Cost of Living Index')
plt.title('Cost of Living Index Across German Cities')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(
    df_students['City'],
    df_students['Apartment (1 bedroom) Outside of Centre']
)
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Monthly Rent (€)')
plt.title('Monthly Rent Outside City Centre for Students')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(
    df_students['City'],
    df_students['Monthly Pass (Regular Price)']
)
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Monthly Transport Cost (€)')
plt.title('Monthly Public Transport Costs Across German Cities')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(
    df_students['City'],
    df_students['Meal, Inexpensive Restaurant']
)
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Meal Cost (€)')
plt.title('Cost of an Inexpensive Meal Across German Cities')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(
    df_students['City'],
    df_students['Cinema, International Release, 1 Seat']
)
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Cinema Ticket Price (€)')
plt.title('Cinema Ticket Prices Across German Cities')
plt.tight_layout()
plt.show()
```

```python
df_students.loc[:, 'Total_Student_Cost'] = (
    df_students['Apartment (1 bedroom) Outside of Centre'] +
    df_students['Meal, Inexpensive Restaurant'] * 30 +
    df_students['Monthly Pass (Regular Price)']
)

df_students[['City', 'Total_Student_Cost']]
```

```python
df_sorted = df_students.sort_values('Total_Student_Cost')

plt.figure(figsize=(10,5))
plt.bar(df_sorted['City'], df_sorted['Total_Student_Cost'])
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Estimated Monthly Student Cost (€)')
plt.title('Overall Student Affordability Across German Cities')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(
    df_students['Apartment (1 bedroom) Outside of Centre'],
    df_students['Total_Student_Cost']
)
plt.xlabel('Monthly Rent Outside Centre (€)')
plt.ylabel('Total Monthly Student Cost (€)')
plt.title('Relationship Between Rent and Total Student Cost')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df_students['Total_Student_Cost'])
plt.ylabel('Total Monthly Student Cost (€)')
plt.title('Distribution of Student Monthly Costs')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(
    df_students['Apartment (1 bedroom) Outside of Centre'],
    df_students['Meal, Inexpensive Restaurant']
)
for i, city in enumerate(df_students['City']):
    plt.text(
        df_students['Apartment (1 bedroom) Outside of Centre'][i],
        df_students['Meal, Inexpensive Restaurant'][i],
        city,
        fontsize=8
    )
plt.xlabel('Monthly Rent (€)')
plt.ylabel('Meal Cost (€)')
plt.title('Rent vs Meal Costs Across German Cities')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Scatter plot of Rent vs Transport
plt.scatter(
    df_students['Apartment (1 bedroom) Outside of Centre'],
    df_students['Monthly Pass (Regular Price)']
)

# Add city labels
for i, city in enumerate(df_students['City']):
    plt.text(
        df_students['Apartment (1 bedroom) Outside of Centre'].iloc[i],
        df_students['Monthly Pass (Regular Price)'].iloc[i],
        city,
        fontsize=8,
        ha='right'  # aligns text neatly
    )

plt.xlabel('Monthly Rent (€)')
plt.ylabel('Monthly Transport (€)')
plt.title('Rent vs Transport Costs Across German Cities')  # fixed complete string
plt.tight_layout()
plt.show()
```
